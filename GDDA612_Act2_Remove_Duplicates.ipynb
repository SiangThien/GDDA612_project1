```python
dfwd = pd.read_csv('olympics_with_dupl.csv', header=1)
dfwd.head()
```

```python
new_names =  {'Unnamed: 0': 'Country',
...               '? Summer': 'Summer Olympics',
...               '01 !': 'Gold',
...               '02 !': 'Silver',
...               '03 !': 'Bronze',
...               '? Winter': 'Winter Olympics',
...               '01 !.1': 'Gold.1',
...               '02 !.1': 'Silver.1',
...               '03 !.1': 'Bronze.1',
...               '? Games': '# Games',
...               '01 !.2': 'Gold.2',
...               '02 !.2': 'Silver.2',
...               '03 !.2': 'Bronze.2'}
```

```python
dfwd.rename(columns=new_names, inplace=True)
```

```python
dfwd2 = dfwd.set_index('Country')
```

```python
# Identify and remove any duplicate records from the dataset

dfwd2.index.duplicated()
```

```python
print(dfwd2[dfwd2.duplicated()])
```

```python
print(dfwd2.duplicated(keep='last'))
```

```python
print(dfwd2.duplicated().sum())
```

```python
print(dfwd2.duplicated().value_counts())
```

```python
print(dfwd2.duplicated(keep=False).value_counts())
```

```python
print(dfwd2.drop_duplicates())
```
