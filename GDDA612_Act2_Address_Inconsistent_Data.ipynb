```python
regex = r'^(\d{4})'
```

```python
extr = dfbl['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
dfbl['Date of Publication'] = pd.to_numeric(extr)
dfbl['Date of Publication'].dtype
```

```python
dfbl['Date of Publication'].isnull().sum() / len(dfbl)
```

```python
dfbl['Place of Publication'].head(10)
```

```python
pub = dfbl['Place of Publication']
london = pub.str.contains('London')
london[:5]
```

```python
oxford = pub.str.contains('Oxford')
```

```python
dfbl['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                               pub.str.replace('-', ' ')))

dfbl['Place of Publication'].head()
```
