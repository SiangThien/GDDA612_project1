```python
dfwd = pd.read_csv('olympics_with_dupl.csv', header=1)
dfwd.head()
```

```python
new_names =  {'Unnamed: 0': 'Country',
...               '? Summer': 'Summer Olympics',
...               '01 !': 'Gold',
...               '02 !': 'Silver',
...               '03 !': 'Bronze',
...               '? Winter': 'Winter Olympics',
...               '01 !.1': 'Gold.1',
...               '02 !.1': 'Silver.1',
...               '03 !.1': 'Bronze.1',
...               '? Games': '# Games',
...               '01 !.2': 'Gold.2',
...               '02 !.2': 'Silver.2',
...               '03 !.2': 'Bronze.2'}
```

```python
dfwd.rename(columns=new_names, inplace=True)
```

```python
# Traverse data frame to detect data types
for column in dfwd.columns:
    print(column,':',pd.api.types.infer_dtype(dfwd[column]))
```

```python
# Transforming mixed data types to single data type
dfwd[column] = dfwd[column].apply(lambda x: pd.to_numeric(x, errors = 'ignore'))
```

```python
# Traverse data frame to detect data types after fix
for column in dfwd.columns:
    print(column, ':', pd.api.types.infer_dtype(dfwd[column]))
```
